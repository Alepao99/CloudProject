```python
import pandas as pd
df = pd.read_csv('continents2.csv' , usecols = ['name','region'])
df
```

```python
d = df.groupby('name')['region'].apply(list).to_dict()
```

```python
d
```

```python
d['Dominican'] = ['Americas']
d['Saudi'] = ['Asia']
d['United States of America'] = ['Americas']
d['North Korea'] = ['Asia']
d['Republic of the Congo'] = ['Africa']
d['Ivory Coast'] = ['Africa']
d['United Kingdom'] = ['Europe']
d['United Republic of Tanzania'] = ['Africa']
```

```python
for file in os.listdir('data/Country/'):
    with open('data/Country/' + file) as f:
        information = True
        for line in f.readlines():
            if information:
                rows = line.strip().split(' ')
                if (rows[0] == '%%' and rows[3] == 'Country:'): #Stiamo nell'information
                    if rows[4] not in d:
                        print(rows)
                        break
                        
```

```python
df_result = pd.DataFrame(columns=('Region', 'Country' ,'CityCode', 'City', 'Month', 'Year', 'AvgTemp'))
CityCode = 0
i = 0
for file in os.listdir('data/Country/'):
    with open('data/Country/' + file) as f:
        information = True
        header = False
        body = False

        for line in f.readlines():
            if information:
                rows = line.strip().split(' ')
                if (rows[0] == '%%' and rows[3] == 'Country:'): #Stiamo nell'information
                    if len(rows) == 5:
                        Country = rows[4]
                    elif len(rows) == 6:
                        Country = rows[4] + ' ' + rows[5]
                    elif len(rows) == 8:
                        Country = rows[4] + ' ' + rows[5] + ' ' + rows[6] + ' ' + rows[7]
                if(rows[0] == '%%' and rows[4] == 'Cities:'):
                    City = rows[5][:-1]
                    information = False
                    header = True
                
            if header:
                if line.strip().split() == []:
                    header = False
                    body = True
            
            if body:
                if line.strip().split(' ') == ['']:
                    pass
                else:
                    data = re.split(r"\s{1,}", line.strip())
                    df_result.loc[i] = [d[Country][0], Country, CityCode, City, data[1], data[0], data[2]]
                    i = i+1
    CityCode = CityCode + 1
```

```python
df_result.info()
```

```python
df_result.to_csv('DatasetCsv/database.csv', index = False)
```

```python
for file in os.listdir('data/'):
    with open('data/' + file) as f:
        header = True
        i = 0
        for line in f.readlines():
            if header:
                header = False
                head = line.strip().split(',')
                df_result = pd.DataFrame(columns=('Year', 'Month' ,'AverageTemperature'))
                del head
            else:
                data = re.split(r"\s{2,}", line.strip())
                if data == ['']:
                    pass
                else:
                    df_result.loc[i] = [data[0], data[1], data[2]]
                    i = i+1

    df_result.to_csv('DatasetCsv/' +file.split('.')[0] + '.csv', index = False)
    break
    del df_result 
```

```python
for file in os.listdir('data/'):
    with open('data/' + file) as f:
        header = True
        i = 0
        for line in f.readlines():
            if header:
                header = False
                head = line.strip().split(',')
                df_result = pd.DataFrame(columns=('dt', 'AverageTemperature', 'Outcome'))
                del head
            else:
                data = re.split(r"\s{2,}", line.strip())
                if data == ['']:
                    pass
                else:
                    if data[0] >= '2016':
                        df_result.loc[i] = [data[0] + '-' + data[1], data[2], 1]
                        i = i+1
                    else:
                        df_result.loc[i] = [data[0] + '-' + data[1], data[2], 0]
                        i = i+1
    df_result.to_csv('DatasetCsvOutcome/' +file.split('.')[0] + '.csv', index = False)
    break
    del df_result 
```
